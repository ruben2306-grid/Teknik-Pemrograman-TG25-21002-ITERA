```python
bilangan = 22.9 # nilai variabelnya bisa diganti

if bilangan % 2 == 1:
  print("bilangan ganil")
elif bilangan % 2 == 0:
  print("bilangan genap")
else:
  print("bukan bilangan")
```

```python
# Masukkan bilangan
Nilai = 50 # nilai variabelnya bisa diganti

if Nilai >= 90:
  print("Nilai A")
elif Nilai >= 80:
  print("Nilai AB")
elif Nilai >= 70:
  print("Nilai B")
elif Nilai >= 60:
  print("Nilai BC")
else:
  print("Nilai C")
```

```python
# Check harga belanja (dengan if)
belanja = 100000 # Nilai variabelnya bisa diganti
pake_kupon = True # Bisa diganti True dan False

if belanja >= 100000:
  if pake_kupon:
    kupon = belanja * 0.2
    akhir = belanja - kupon
    print("Pake kupon,",int(akhir))
  else:
    print("Gak pake kupon,",int(belanja))
else:
  print("uang belanja tidak cukup")
```

```python
# Perulangan For, While, do-While
for i in range(1, 101, 1): # range(start, end, space)
  i **= 2
  print(i)
```

```python
for i in range(0,100,1):
  if i % 2 == 0:
    print(f"Bil {i} adalah genap")
  elif i % 2 != 0:
    print(f"Bil {i} adalah ganjil")
  else:
    print(f"Bil {i} adalah bukan bilangan bulat")
```

```python
# Penggunaan while
i = 0
while i >= 0:
  i = i + 1
  print(i)

  if i >= 100:
    break
```

```python
# Tugas / Challenge 1
# Simulasi Program Login
# versi buatan sendiri
Percobaan = 3
username = "admin"
password = "admin123"
login = True

while login:
    username_1 = input("masukkan input: ")
    password_1 = input("masukkan password: ")

    Percobaan += 1
    if username == username_1 and password == password_1:
      print("login berhasil")
      print(username_1)
      print(password_1)
      break #bisa juga ganti "login = False", intinya berhenti juga
    else:
      print("keduanya salah")
  
    if Percobaan == 3:
      break 
```

```python
# Tugas / Challenge 2 
# Rata-rata usia total anggota keluarga
total_anggota_keluarga = int(input("Jumlah anggota keluarga: "))
total_keseluruhan = 0

for jumlah in range(total_anggota_keluarga):
  usia = int(input(f"Usia anggota ke-{jumlah + 1}: "))
  total_keseluruhan += usia

rata_rata = total_keseluruhan / total_anggota_keluarga
print("Rata-rata usia:", round(rata_rata, 2))
```
